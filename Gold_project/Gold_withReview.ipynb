### –ü—Ä–æ–≤–µ—Ä–∫–∞ —Ä–∞—Å—á—ë—Ç–æ–≤ —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏ –æ–±–æ–≥–∞—â–µ–Ω–∏—è

–ü—Ä–æ–≤–µ—Ä–∏–º, —á—Ç–æ —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç—å –æ–±–æ–≥–∞—â–µ–Ω–∏—è —Ä–∞—Å—Å—á–∏—Ç–∞–Ω–∞ –ø—Ä–∞–≤–∏–ª—å–Ω–æ.
- –í—ã—á–∏—Å–ª–∏–º recovery –Ω–∞ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–µ train –¥–ª—è –ø—Ä–∏–∑–Ω–∞–∫–∞ rougher.output.recovery (—ç—Ç–∞–ø —Ñ–ª–æ—Ç–∞—Ü–∏–∏).
- –ù–∞–π–¥—ë–º MAE –º–µ–∂–¥—É —Ä–∞—Å—á—ë—Ç–∞–º–∏ –∏ –∑–Ω–∞—á–µ–Ω–∏–µ–º –ø—Ä–∏–∑–Ω–∞–∫–∞.

```python
# —Ä–∞—Å—á—ë—Ç —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏ –æ–±–æ–≥–∞—â–µ–Ω–∏—è –∑–æ–ª–æ—Ç–∞ –Ω–∞ —ç—Ç–∞–ø–µ —Ñ–ª–æ—Ç–∞—Ü–∏–∏
C = data_train.loc[~data_train['rougher.output.recovery'].isna(), 'rougher.output.concentrate_au']
F = data_train.loc[~data_train['rougher.output.recovery'].isna(), 'rougher.input.feed_au']
T = data_train.loc[~data_train['rougher.output.recovery'].isna(), 'rougher.output.tail_au']
rough_recov_calc = C * (F - T) / (F * (C - T)) * 100
rough_recov_train = data_train.loc[~data_train['rougher.output.recovery'].isna(), 'rougher.output.recovery']
```

```python
# —Ä–∞—Å—á—ë—Ç sMAPE 
numer = abs(rough_recov_train - rough_recov_calc)
denom = 0.5 * (abs(rough_recov_train) + abs(rough_recov_calc))
N = len(rough_recov_train)
smae_rough_rec = 1 / N * (numer / denom).sum() * 100
print("–ü–æ–∫–∞–∑–∞—Ç–µ–ª—å sMAPE –¥–ª—è '–≠—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏ –æ–±–æ–≥–∞—â–µ–Ω–∏—è –∑–æ–ª–æ—Ç–∞ –Ω–∞ —ç—Ç–∞–ø–µ —Ñ–ª–æ—Ç–∞—Ü–∏–∏'")
print("–º–µ–∂–¥—É —Ä–∞—Å—Å—á–∏—Ç–∞–Ω–Ω—ã–º –∑–Ω–∞—á–µ–Ω–∏–µ–º –∏ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–æ–π –≤—ã–±–æ—Ä–∫–æ–π: {0:.2E}".format(smae_rough_rec))
```

- –û–ø—Ä–µ–¥–µ–ª–∏–º —É—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ alpha = 0.01
- –î–ª—è —Ä–∞—Å—á—ë—Ç–∞ —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏ —Ä–∞–∑–Ω–æ—Å—Ç–∏ –∏ p-value –±—É–¥–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é st.ttest_ind

–ó–Ω–∞—á–µ–Ω–∏–µ p-value –Ω–∏–∂–µ 0.01, –ø–æ–ø—Ä–æ–±—É–µ–º —É–±—Ä–∞—Ç—å –≤—ã–±—Ä–æ—Å—ã –≤ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è—Ö –∏ –ø—Ä–æ–≤–µ—Å—Ç–∏ t-test –µ—â—ë —Ä–∞–∑.

```python
train_q_75 = data_train['rougher.input.feed_size'].quantile(0.75)
train_q_25 = data_train['rougher.input.feed_size'].quantile(0.25)
test_q_75 = data_test['rougher.input.feed_size'].quantile(0.75)
test_q_25 = data_test['rougher.input.feed_size'].quantile(0.25)
train_min = train_q_25 - 3 * (train_q_75 - train_q_25)
train_max = train_q_75 + 3 * (train_q_75 - train_q_25)
test_min = test_q_25 - 3 * (test_q_75 - test_q_25)
test_max = test_q_75 + 3 * (test_q_75 - test_q_25)
print('–¢—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–∞—è –≤—ã–±–æ—Ä–∫–∞:')
print('    –ù–∏–∂–Ω—è—è –≥—Ä–∞–Ω–∏—Ü–∞ –≤—ã–±—Ä–æ—Å–æ–≤: {:.2f}'.format(train_min))
print('    –í–µ—Ä—Ö–Ω—è—è –≥—Ä–∞–Ω–∏—Ü–∞ –≤—ã–±—Ä–æ—Å–æ–≤: {:.2f}'.format(train_max))
print('–¢–µ—Å—Ç–æ–≤–∞—è –≤—ã–±–æ—Ä–∫–∞:')
print('    –ù–∏–∂–Ω—è—è –≥—Ä–∞–Ω–∏—Ü–∞ –≤—ã–±—Ä–æ—Å–æ–≤: {:.2f}'.format(test_min))
print('    –í–µ—Ä—Ö–Ω—è—è –≥—Ä–∞–Ω–∏—Ü–∞ –≤—ã–±—Ä–æ—Å–æ–≤: {:.2f}'.format(test_max))
```

```python
results = st.ttest_ind(train_feed_size.loc[(train_feed_size > train_min) &
                                           (train_feed_size < train_max)],
                       test_feed_size.loc[(test_feed_size > test_min) &
                                          (test_feed_size > test_min)],
                       equal_var=True)
print('p-value: {:.3f}'.format(results.pvalue))
```

**–í—ã–≤–æ–¥:** p-value –≤—ã—à–µ 0.01, —á—Ç–æ –≥–æ–≤–æ—Ä–∏—Ç –æ —Ç–æ–º, —á—Ç–æ –≤—ã–±–æ—Ä–∫–∏ –Ω–µ —Å–∏–ª—å–Ω–æ –æ—Ç–ª–∏—á–∞—é—Ç—Å—è –¥—Ä—É–≥ –æ—Ç –¥—Ä—É–≥–∞ –∏ –ø—Ä–∏–∑–Ω–∞–∫ "–†–∞–∑–º–µ—Ä –≥—Ä–∞–Ω—É–ª —Å—ã—Ä—å—è" –º–æ–∂–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –¥–ª—è –æ–±—É—á–µ–Ω–∏—è –º–æ–¥–µ–ª–∏.

```python
stages = ['rougher.input.feed', 'rougher.output.concentrate', 'final.output.concentrate']

substance = ['au', 'pb', 'sol', 'ag']
```

```python
# —Ñ—É–Ω–∫—Ü–∏—è, –≤–æ–∑–≤—Ä–∞—â–∞—é—â–∞—è —Å–ø–∏—Å–æ–∫ —Å—Ç–æ–ª–±—Ü–æ–≤ –¥–ª—è –∫–∞–∂–¥–æ–π —Å—Ç–∞–¥–∏–∏
def get_column_list(stage, substance):
    return list(map(lambda item: stage + '_' + item, substance))
```

```python
# –¥–∞—Ç–∞—Ñ—Ä–µ–π–º —Å —Å—É–º–º–∞–º–∏ –¥–ª—è –∫–∞–∂–¥–æ–π —Å—Ç–∞–¥–∏–∏
def get_substance_sum(data_train):
    substance_sum = pd.DataFrame()

    for stage in stages:
        substance_sum[stage] = data_train[get_column_list(stage, substance)].sum(axis=1)
    return substance_sum
```

```python
def hampel(vals_orig):
    vals = vals_orig.copy()    
    difference = np.abs(vals.median()-vals)
    median_abs_deviation = difference.median()
    threshold = 5 * median_abs_deviation
    outlier_idx = difference > threshold
    vals[outlier_idx] = np.nan
    return(vals)
```

### –§–æ—Ä–º—É–ª—ã –¥–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –º–µ—Ç—Ä–∏–∫ –∫–∞—á–µ—Å—Ç–≤–∞

```python
def sMAPE(target, prediction):
    return 1 / len(target) * sum( abs(target - prediction) / ( (abs(target) + abs(prediction)) * 0.5) ) * 100
```

–¢–∞–∫ –∫–∞–∫ —Å—Ç–æ–ª–±—Ü—ã –≤ —Ç–µ—Å—Ç–æ–≤–æ–π –∏ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–∞—Ö —Ä–∞–∑–ª–∏—á–∞—é—Ç—Å—è, –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ —Å—Ñ–æ—Ä–º–∏—Ä–æ–≤–∞—Ç—å –Ω–æ–≤—É—é –æ–±—É—á–∞—é—â—É—é –≤—ã–±–æ—Ä–∫—É —Ç–æ–ª—å–∫–æ —Å —Ç–µ–º–∏ —Å—Ç–æ–ª–±—Ü–∞–º–∏, –∫–æ—Ç–æ—Ä—ã–µ –µ—Å—Ç—å –≤ —Ç–µ—Å—Ç–æ–≤–æ–π.

```python
data_train_full = data_train.copy()

data_train = data_train.drop(list(data_train.columns.difference(data_test.columns)), axis=1)
```

–ò—Ç–æ–≥–æ–≤–∞—è –º–µ—Ç—Ä–∏–∫–∞ sMAPE –∑–∞–≤–∏—Å–∏—Ç –ª–∏–Ω–µ–π–Ω–æ –æ—Ç sMAPE –ø–æ—Å–ª–µ —Ñ–ª–æ—Ç–∞—Ü–∏–∏ –∏ –ø–æ—Å–ª–µ –æ—á–∏—Å—Ç–∫–∏, –∑–Ω–∞—á–∏—Ç, –æ–±—É—á–∞—Ç—å –º–æ–¥–µ–ª—å –º–æ–∂–Ω–æ –Ω–∞ –¥–≤—É—Ö –Ω–µ–∑–∞–≤–∏—Å–∏–º—ã—Ö –≤—ã–±–æ—Ä–∫–∞—Ö. 

–í—ã–¥–µ–ª–∏–º –≤ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ –∏ —Ü–µ–ª–µ–≤—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏.

–°—Ñ–æ—Ä–º–∏—Ä—É–µ–º —Å–ø–∏—Å–∫–∏ —Å—Ç–æ–ª–±—Ü–æ–≤ –¥–ª—è rougher –∏ final –≤—ã–±–æ—Ä–æ–∫.

```python
rougher_columns = []
pattern = re.compile('rougher')

for column in list(data_test.columns):
    if re.match(pattern, column):
        rougher_columns.append(column)

final_columns = list(set(data_test.columns) - set(rougher_columns))
```

–°–∫–æ–ø–∏—Ä—É–µ–º –∏–∑ –æ–±—â–µ–π –≤—ã–±–æ—Ä–∫–∏ –≤ —Ç–µ—Å—Ç–æ–≤—É—é –∑–Ω–∞—á–µ–Ω–∏—è –∏–∑ —Ü–µ–ª–µ–≤—ã—Ö —Å—Ç–æ–ª–±—Ü–æ–≤ –ø–æ –∏–Ω–¥–µ–∫—Å—É —Å—Ç—Ä–æ–∫

```python
test_index = list(data_test.index)

test_target_columns = pd.DataFrame()
test_target_columns = data_full.loc[test_index][['rougher.output.recovery', 'final.output.recovery']]
```

<div class="alert alert-success">
<h2> –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞ <a class="tocSkip"> </h2>

<b>–í—Å–µ –æ—Ç–ª–∏—á–Ω–æ!üëç:</b> 
    
–í—ã—Ç—è–≥–∏–≤–∞–Ω–∏–µ —Ü–µ–ª–µ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –≤ —Ç–µ—Å—Ç–æ–≤—É—é –≤—ã–±–æ—Ä–∫—É –ø—Ä–∞–≤–∏–ª—å–Ω–æ–µ) –ü—Ä–∏–≤—è–∑–∫–∞ –∫ –¥–∞—Ç–µ - –≤–µ—Ä–Ω–æ–µ —Ä–µ—à–µ–Ω–∏–µ

–†–∞–∑–¥–µ–ª–∏–º –≤—ã–±–æ—Ä–∫–∏

```python
rougher_train_features = data_train[rougher_columns]
rougher_train_target = data_train_full['rougher.output.recovery']

final_train_features = data_train[final_columns]
final_train_target = data_train_full['final.output.recovery']

rougher_test_features = data_test[rougher_columns]
rougher_test_target = test_target_columns['rougher.output.recovery']

final_test_features = data_test[final_columns]
final_test_target = test_target_columns['final.output.recovery']
```

–ü—Ä–æ–≤–µ—Ä–∏–º —Ä–∞–∑–º–µ—Ä—ã —Å—Ñ–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –≤—ã–±–æ—Ä–æ–∫

```python
print('–†–∞–∑–º–µ—Ä—ã rougher –≤—ã–±–æ—Ä–∫–∏', rougher_train_features.shape, rougher_train_target.shape)
print('–†–∞–∑–º–µ—Ä—ã final –≤—ã–±–æ—Ä–∫–∏', final_train_features.shape, final_train_target.shape)
print('–†–∞–∑–º–µ—Ä—ã rougher test –≤—ã–±–æ—Ä–∫–∏', rougher_test_features.shape, final_test_features.shape)
```

–ú–∞—Å—à—Ç–∞–±–∏—Ä—É–µ–º –ø—Ä–∏–∑–Ω–∞–∫–∏ –Ω–∞ –æ–±—É—á–∞—é—â–µ–π –∏ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–∞—Ö —Å –ø–æ–º–æ—â—å—é –º–µ—Ç–æ–¥–∞ StandardScaler

```python
def fit_scaler(df):
    scaler = StandardScaler()
    scaler.fit(df)
    return scaler
def apply_scaler(df, scaler):
    return pd.DataFrame(scaler.transform(df))
```

–û–±—É—á–∏–º Scaler –Ω–∞ –æ–±–µ–∏—Ö –æ–±—É—á–∞—é—â–∏—Ö –≤—ã–±–æ—Ä–∫–∞—Ö (rougher –∏ final)

```python
rougher_scaler = fit_scaler(rougher_train_features)
final_scaler = fit_scaler(final_train_features)
```

–ü—Ä–∏–º–µ–Ω–∏–º –æ–±—É—á–µ–Ω–Ω—ã–π Scaler –∫–æ –≤—Å–µ–º –≤—ã–±–æ—Ä–∫–∞–º

```python
rougher_train_features = apply_scaler(rougher_train_features, rougher_scaler)
rougher_test_features = apply_scaler(rougher_test_features, rougher_scaler)
```

```python
final_train_features = apply_scaler(final_train_features, final_scaler)
final_test_features = apply_scaler(final_test_features, final_scaler)
```

### –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–µ–π

–°–æ–∑–¥–∞–¥–∏–º –º–æ–¥–µ–ª–∏ RandomForestRegressor –∏ DecisionTreeRegressor

```python
model_rf = RandomForestRegressor()
model_dt = DecisionTreeRegressor()
```

–ó–∞–¥–∞–¥–∏–º –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –¥–ª—è GridSearchCV

```python
parameters_rf = {"n_estimators":[5,100], "max_depth":[1,10]}
parameters_dt = {"max_depth":[1,10]}
```

```python
def get_grid(model, parameters, custom_score, features, target):
    grid = GridSearchCV(model, parameters, custom_score)
    grid.fit(features, target)
    return grid
```

```python
grid_rougher_rf = get_grid(model_rf, parameters_rf, custom_score, rougher_train_features, rougher_train_target)
grid_rougher_dt = get_grid(model_dt, parameters_dt, custom_score, rougher_train_features, rougher_train_target)

grid_final_rf = get_grid(model_rf, parameters_rf, custom_score, final_train_features, final_train_target)
grid_final_dt = get_grid(model_dt, parameters_dt, custom_score, final_train_features, final_train_target)
```

```python
def display_grid(grid):
    print('Best score:', grid.best_score_, 
          '\nBest estimator:', grid.best_estimator_)
```

```python
display('RF roughter', display_grid(grid_rougher_rf))
display('DT roughter', display_grid(grid_rougher_dt))
display('RF final', display_grid(grid_final_rf))
display('DT final', display_grid(grid_final_dt))
```

```python
sMAPE_final(grid_rougher_rf.best_score_, grid_final_rf.best_score_)
```
